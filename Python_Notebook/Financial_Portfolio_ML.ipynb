```python
import yfinance as yf
ticker_symbol = ["GC=F", "SI=F", "BTC-USD", "SPY", "VBIAX","VIX", "VXN", "TNX", "TLT", "XLK", "XLF", "XLV","XLI", "QQQ"]
data = yf.download(ticker_symbol, period="20y", interval="1d")
close_price = data['Close']
print("Daily Closing Prices for Interest Assets (Last 52 Weeks):")
print(close_price.tail())
```

```python
print("Daily Closing Prices for Interest Assets (First 52 Weeks):")
print(close_price.head())
```

```python
print("Daily Closing Prices for Interest Assets (Sample rows from the 52 Weeks):")
print(close_price.sample(5))
```

```python
close_price.to_csv("daily_closing_prices_full.csv", index=False)
```
