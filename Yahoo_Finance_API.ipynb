```python
import pandas as pd
import yfinance as yf
from google.colab import files

# Configuration matching your diagram metadata
tickers = ["GC=F", "SI=F", "BTC-USD", "SPY", "VBIAX"]
asset_info = [
    {"id": 1, "ticker": "BTC-USD", "name": "Bitcoin", "type_id": 1, "type_name": "Cryptocurrency"},
    {"id": 2, "ticker": "GC=F", "name": "Gold Futures", "type_id": 2, "type_name": "Commodity"},
    {"id": 3, "ticker": "SI=F", "name": "Silver Futures", "type_id": 2, "type_name": "Commodity"},
    {"id": 4, "ticker": "SPY", "name": "SPDR S&P 500 ETF", "type_id": 3, "type_name": "ETF"},
    {"id": 5, "ticker": "VBIAX", "name": "Vanguard Balanced Index", "type_id": 4, "type_name": "Mutual Fund"}
]
source_id = 101
sql_file = "relational_stock_data.sql"

# Download historical data
print("Downloading 1 year of historical data...")
data = yf.download(tickers, period="1y", interval="1d")

with open(sql_file, "w") as f:
    # 1. DROP existing tables in reverse dependency order
    f.write("IF OBJECT_ID('Prices', 'U') IS NOT NULL DROP TABLE Prices;\n")
    f.write("IF OBJECT_ID('Asset', 'U') IS NOT NULL DROP TABLE Asset;\n")
    f.write("IF OBJECT_ID('Asset_Type', 'U') IS NOT NULL DROP TABLE Asset_Type;\n")
    f.write("IF OBJECT_ID('Data_Source', 'U') IS NOT NULL DROP TABLE Data_Source;\n\n")

    # 2. CREATE tables exactly as per Crow's Foot diagram
    f.write("CREATE TABLE Asset_Type (Asset_Type_ID INT PRIMARY KEY, Type_Name VARCHAR(50));\n")
    f.write("CREATE TABLE Asset (Asset_ID INT PRIMARY KEY, Ticker_Symbol VARCHAR(20), Asset_Name VARCHAR(100), Asset_Type_ID INT FOREIGN KEY REFERENCES Asset_Type(Asset_Type_ID));\n")
    f.write("CREATE TABLE Data_Source (Data_Source_ID INT PRIMARY KEY, Source_Name VARCHAR(100), Last_Sync_Time_Stamp DATETIME);\n")
    f.write("CREATE TABLE Prices (Price_ID INT IDENTITY(1,1) PRIMARY KEY, Asset_ID INT FOREIGN KEY REFERENCES Asset(Asset_ID), Price_Date DATE, Close_Price DECIMAL(18, 2), Volume BIGINT, Data_Source_ID INT FOREIGN KEY REFERENCES Data_Source(Data_Source_ID));\n\n")

    # 3. INSERT Seed Data (Metadata)
    types_added = set()
    for a in asset_info:
        if a['type_id'] not in types_added:
            f.write(f"INSERT INTO Asset_Type VALUES ({a['type_id']}, '{a['type_name']}');\n")
            types_added.add(a['type_id'])

    f.write(f"INSERT INTO Data_Source VALUES ({source_id}, 'Yahoo Finance API', GETDATE());\n")

    for a in asset_info:
        f.write(f"INSERT INTO Asset VALUES ({a['id']}, '{a['ticker']}', '{a['name']}', {a['type_id']});\n")
    f.write("\n")

    # 4. INSERT Time Series Data (Prices)
    for date, row in data['Close'].iterrows():
        d_str = date.strftime('%Y-%m-%d')
        for a in asset_info:
            price = row[a['ticker']]
            volume = data['Volume'].loc[date, a['ticker']]
            if pd.notnull(price):
                v_val = int(volume) if pd.notnull(volume) else 0
                f.write(f"INSERT INTO Prices (Asset_ID, Price_Date, Close_Price, Volume, Data_Source_ID) VALUES ({a['id']}, '{d_str}', {round(float(price), 2)}, {v_val}, {source_id});\n")

print(f"Relational SQL file '{sql_file}' generated.")
files.download(sql_file)
```
